```python
import fastai.basics as fai
import fastai.vision as fv
import numpy
from pathlib import Path
from shutil import copyfile
import random
import torch
import torch.nn.functional as F
import torch.nn as nn
import PIL
from PIL import Image, ImageDraw, ImageFont
from tqdm import tqdm_notebook as tqdm
from Sketch import *
```

```python
path_good   = Path("faces")
path_crappy = Path("sketches")
```

```python
path_crappy.mkdir(parents=True, exist_ok=True)
```

```python
def crappy_name(good_name):
    good_name = good_name.relative_to(path_good)
    crappy_file = path_crappy/good_name
    crappy_file.parent.mkdir(parents=True, exist_ok=True)
    return crappy_file


def crappify(file,sigma):
    crappy_file = crappy_name(file)
    
    img = PIL.Image.open(file)
    
    sketch = sketchify(np.array(img), sigma)
    sketch = numpy.array(sketch, dtype = np.float64)
    
    sketch -= sketch.min()
    sketch *= (255.0/sketch.max())
    
    sketch = numpy.array(sketch, dtype = np.uint8)
    sketch = Image.fromarray(sketch)
    
    sketch.save(crappy_file)
```

```python
all_files = fv.get_image_files(path_good,recurse=True)
```

```python
sigma = 8
for file in all_files:
    crappify(file, sigma)
```
